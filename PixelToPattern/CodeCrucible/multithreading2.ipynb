#  Semaphores to avoid the issue of race condition 
    Lock and RLock , only one thread is allowed to execute 
    but sometimes our requirement is to execute a particular number of threads at a time then use semaphores

    # real world eg :
        university : website : result 
        if there are n thread then n student can access the result simultaneously 

Semaphore can be used to limit the access to the shared resources with limited capacity 

# create an object of semaphores class 


```python

# object of sempaphore -> acquire -> code -> release 
# object banate samay total count dena padta hain , jaise hi koi thread acquire karta hai 
# fir total count ki value ek se -- kar di jati hai aur jaise hi thread release karta hain waise hi total count ++ ho jata hai 
# jab total count 0 hoga tab baaki thread ko wait karna padega critical section ko access karne ke liye 
#  

from threading import * 
import time 
# when argument is not given , it behave like Lock and Rlock 
# by default value is one 
obj = Semaphore(3) # only three threads are allowed at a time 
def display(name) : 
    obj.acquire() 
    for i in range(5) : 
        print("Namaste Duniya") 
        print(name) 
        time.sleep(0.78) 
    obj.release() 


t1 =Thread(target=display , args =('Thread - 1',))
t2 =Thread(target=display , args =('Thread - 2',))
t3 =Thread(target=display , args =('Thread - 3',))
t4 =Thread(target=display , args =('Thread - 4',))
t5 =Thread(target=display , args =('Thread - 5',))

# args must be tuple for single element 

t1.start( ) 
t2.start() 
t3.start() 
t4.start() 
t5.start() 

# ek baar acquire karne par aur do baar release karne par issue hp jayega 
# to avoid this issue use BoundedSemaphore () 
# best practise is to use BoundendSemphore () 

```

```python
# using event 
# a flag variable is mainted , in starting it is False 
# wait() method , 
# steps 
# create an event object 
# create two threads which will communicate 
# put t2 thread in wiating by using wait() 
# use set() method in/after t1 threads code

# set() : event class ka method hain 
# set the internal flag to true
# if the flag is True , waiting thread is awakened 
# reset() : 
    # reset the internal flag to false 
    # other thread will wait() again 
# is_set() :  return bool value, check whethe internal value is true or not 

##wait([timeout] ) : calling the function keep other thread on wait until flag is set to True 
```

```python
import threading
import  time  
e = threading.Event() 
def task() :
    print("Game is started ") 
    time.sleep(3) 
    e.set() # flag = True 
def end() : 
    e.wait() # t2 thread will wait, tab tak wait karo jab tak signal nahi mitla hain 
    if(e.is_set()) :
        print("Destoying the session") 

t1 = threading.Thread(target=task) 
t2 = threading.Thread(target=end) 
t1.start() 
t2.start() 

```

```python
# thread communication using file handling 
import threading 
from time import sleep 

e = threading.Event() # calling constructor of Event class 
def light_switch() :
    while(True) : 
        print("Light is Green") 
        e.set() 
        sleep(5) 
        print("Light is red") 
        e.clear() 
        sleep(5) 
        e.set() 


def traffic_message() :
    e.wait() 
    while(e.is_set()) :
        print("You can go!") 
        sleep(0.5) # light blinking
        # fir se wait state me bhejo 
        e.wait() 

t1= threading.Thread(target=light_switch)
t2 = threading.Thread(target= traffic_message) 
t1.start() 
t2.start() 

# wait(timeout = 0) # timeout ka default value hain 0 

```

```python
# communication b/w multiple threads
# two threads -> Event class 
# Condition objects
# need : 
 # to communicate with multiple threads(producer and consumer ) 

# notify() : wakeup only one thread 
# notify_all() : wakeup multiple thread 
#wait(timeout =0 ) # Condition Object 

import threading 
import time 

def write_data() :
    con.acquire() 
    with open('report.txt' ,'a') as file :
        days = ['Monday' , 'Tuesday' , 'Wednesday' , 'Thursday' ,'Friday' , 'Saturday' , 'Sunday'] 
        for day in days :
            temp = input("Enter the temperature for {}".format(day)) 
            file.write(temp+ "\n") 
    con.notify_all() 

    con.release()
def max_temp() :
    con.acquire() 
    con.wait(timeout=0) 
    with open("report.txt" ,"r") as file :
        data = file.readlines() 
        max1 = float(data[0].strip("\n")) 
        for temp in data[1:] :
            temp = float(temp.strip("\n")) 
            if (temp>max1) :
                max1 = temp 
        print("Maximum temperature is : ", max1)
    con.release()
def avg_temp() : 
    con.acquire() 
    con.wait(timeout = 0) 
    with open("report.txt" ,"r")as file : 
        data = file.readlines() 
        sum = 0 
        for temp in data : 
            temp = float(temp.strip("\n")) 
            sum += temp 

        avg = sum/len(data) 
        print("Average temperature of week is {}".format(avg))
    con.release()
# condition object 
con = threading.Condition()
t1 = threading.Thread(target= write_data) 
t2 = threading.Thread(target= max_temp) 
t3 = threading.Thread(target= avg_temp) 

t1.start( )
t2.start() 
t3.start() 

```

```python
# # queue module ke andar Queue class hai 
# # syntax : 
#     # import queue 
#     # q = queue.Queue(size) size is optional 
# # put(item , block = True) : value to inset into queue, False karne se exception aayega jab size se adhik element rakhenge to 
# # get() : used to delete an item from queue and return the delete element 
# # Thread safe : No race condition 
# # implements all the required locking semantics 

from time import sleep 
from threading import Thread 
from queue import Queue 

def producer(q) :
    print("Producer is running :") 
    n = int(input("Enter the number of students : ") )
    for i in range(n) :
        marks = eval(input("Enter the marks : ")) 
        q.put(marks) 
    q.put(None) 

def consumer(q):
    print("Consumer is running : ") 
    while(True) : 
        item = q.get() 
        if(item is None) :
            break 
        print(f"Got item {item}")
    print("Consumer End!") 

q = Queue() 
t1 = Thread(target=producer , args=(q,)) 
t2 = Thread(target= consumer , args=(q)) 
t1.start() 
t2.start() 
```

```python
# daemon threads and how to create them..... 
# python idle 
# Main thread 
# thread for interactive shell
# thread for editor 
# Thread for python interpreter 
# Thread for text highlightig 
# Thread for memory management 


# daemon threads : supportive threads 
# non daemon threads : non supportive threads

# programm will not terminate until all non daemon thread gets completed 
# daemon thread , when all non daemon threads get terminated automatically daemon threads also get terminated 
#  
# Daemon Thread : a daemon thread is a thread which runs continuously in background and provide support to other non -daemon threads 

# used for monitoring , background services or cleanup resources 

# what decide a deamon thread nature ? 
# t1.daemon  if true then thread is daemon else non daemon 
# t1.daemon = value 
# we can't change the daemon nature of running threads 

```

```python
# checking daemon nature of main thread : false 
from threading import * 
obj = current_thread() 
print(obj.daemon) # False 
# it means main thread is non daemon thread 

# changing the nature of thread 

def display() :
    print("This is a function") 

t1= Thread(target=display) 
# daemon nature milta hain parent thread ke nature se 
print("Daemon nature of t1 : {}".format(t1.daemon))
# changing its nature 
t1.daemon = True 
print("After changing Daemon nature of t1 : {}".format(t1.daemon))
t1.start() 

# changing daemon nature after starting thread
# 
try : 
    t1.daemon = False # runtime error 
except RuntimeError as err:
    print(err) 
else :
    print("Nature has been changed") 
    
# error 
```

```python
# barrier object 
# What is barrier object and how to create it 
# cycle racing 
# allows threads to wait for each other at a specific point of their execution before proceeding further

import threading 
import time 
number= 4
br = threading.Barrier(number)
def player(name) :
    print(f"{name} started ")
    score = 0 
    for i in range(5) :
        time.sleep(2.8) 
        print(f"{name} is playing") 
        score+= i 
    # barrier 
    br.wait() 
    
    print(f"{name} scored {score}") 
    print(f"Sending wiining amount to {name}") 

Thread = [] 
player_names = ["Brijesh" , "Aman" ,"Atit" , "Ankit"]
for name in player_names :
    thread = threading.Thread(target=player , args=(name,)) 
    Thread.append(thread) 
    thread.start() 


    
```
